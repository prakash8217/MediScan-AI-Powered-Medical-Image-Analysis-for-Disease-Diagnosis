```python
import base64
import io
from PIL import Image

@app.callback(
    Output("upload-image", "children"),
    [Input("upload-image", "contents")]
)
def update_output(image_contents):
    if image_contents is not None:
        # Decode and display the uploaded image
        data = image_contents.encode("utf8").split(b";base64,")[1]
        img = Image.open(io.BytesIO(base64.b64decode(data)))
        img = img.resize((400, 300))  # Resize for display purposes
        return html.Img(src=image_contents, style={"width": "100%", "borderRadius": "10px"})
    return html.Div(["Upload Image", html.Br(), "or drag and drop here"])
```
